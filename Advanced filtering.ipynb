### Graficamos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Datos sacados de tu tabla
labels = ['Mujer', 'Hombre']
noSurvived = [81, 468]
survived = [233, 109]

x = np.arange(len(labels))  # posiciones de las etiquetas (0, 1)
width = 0.35  # ancho de las barras

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 6))

# Dibujar las barras
bars1 = ax.bar(x - width/2, noSurvived, width, label='No sobrevivió', color='orange')
bars2 = ax.bar(x + width/2, survived, width, label='Sobrevivió', color='tomato')

# Etiquetas y título
ax.set_xlabel('Sexo')
ax.set_ylabel('Cantidad de personas')
ax.set_title('Supervivencia por sexo')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Mostrar los valores encima de las barras (opcional pero bonito)
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Desplazamiento del texto
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
