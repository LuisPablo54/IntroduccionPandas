```python
import pandas as pd
```

```python
df = pd.read_csv('13_11_Nov_23_H.csv', encoding='latin1', header=3)
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
df1 = df.drop(['TEMP', 'WD', '03(ppm)'], axis=True)
df1
```

```python
df1.count(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
```

```python
df50 = df1.loc[:50,:]
df50
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df50.index, df50['PM10(µg/m3)'], label='PM10(µg/m3)', color='blue')
ax.plot(df50.index, df50['PM2.5(µg/m3)'], label='PM2.5(µg/m3)', color='green')
ax.set_xlabel('Dias')
ax.set_ylabel('SO2 (ppm)')
ax.set_title('Regristros por dia')
# Se agrega la cuadricula en x
ax.grid(True, axis='y')
ax.legend()
plt.show()

```
