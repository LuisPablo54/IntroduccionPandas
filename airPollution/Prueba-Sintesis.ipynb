# Gráfica

```python
import matplotlib.pyplot as plt
```

```python
T1_50 = T1.loc[:120,:]
CICEG_50 = CICEG.loc[:120,:]
FM_50 = FM.loc[:120,:]
```

## Grafica 1

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(T1_50.index, T1_50['PM10(µg/m3)'], label='Estación T1', color='blue')
ax.plot(CICEG_50['PM10(µg/m3)'], label='Estación CICEG', color='green')
ax.plot(FM_50['PM10 [µg/m³]'], label='Facultad de Medicina', color='red')
ax.axhline(y=45, color='gray', linestyle='--', linewidth=1, label='Límite OMS (24h)')


# Personalizar el gráfico
ax.set_xlabel('Tiempo (120 horas = 5 días)', fontsize=12, labelpad=10)
ax.set_ylabel('Concentración de PM10 [µg/m³]', fontsize=12, labelpad=10)
ax.set_title('Registros horarios de PM10 - Noviembre 2023', fontsize=14, pad=20, weight='bold')



# Se agrega la cuadricula en x
ax.grid(True, axis='y')
ax.legend()
plt.show()

```

```python
fig, axs = plt.subplots(nrows=3, figsize=(8, 5), sharex=True)

# Gráfica 1: Estación T1
axs[0].plot(T1_50['PM10(µg/m3)'], color='blue', label='Estación T1')
axs[0].set_ylabel('PM10(µg/m3)')
axs[0].set_title('Estación T1')
axs[0].set_ylim(0, 160)
axs[0].grid(True)
axs[0].legend()

# Gráfica 2: Estación CICEG
axs[1].plot(CICEG_50['PM10(µg/m3)'], color='green', label='Estación CICEG')
axs[1].set_ylabel('PM10(µg/m3)')
axs[1].set_title('Estación CICEG')
axs[1].set_ylim(0, 160)
axs[1].grid(True)
axs[1].legend()

# Gráfica 3: Facultad de Medicina
axs[2].plot(FM_50['PM10 [µg/m³]'], color='red', label='Facultad de Medicina')
axs[2].set_ylabel('PM10(µg/m3)')
axs[2].set_title('Facultad de Medicina')
axs[2].set_xlabel('Tiempo (120 horas = 5 días)')
axs[2].set_ylim(0, 160)
axs[2].grid(True)
axs[2].legend()

# Título general
fig.suptitle('Registros horarios de NO2 [ppm] - Noviembre 2023', fontsize=10, weight='bold', y=1.02)

plt.tight_layout()
plt.show()
```
