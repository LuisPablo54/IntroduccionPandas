```python
import pandas as pd
```

```python
df = pd.read_csv('todosLosSitios.csv')
df.sample()
```

```python
df.columns
```

```python
nombres = df.sitio.unique()
print(nombres)
```

```python
MayorCalifiacion = df[df.Escala > 4]
```

## Grafica de pastesl

```python
import matplotlib.pyplot as plt
```

```python
# Contar cuántas veces aparece cada sitio con calificación 5
conteoOpiniones = MayorCalifiacion['sitio'].value_counts()

# Crear gráfica de pastel
plt.figure(figsize=(8, 10))
plt.pie(conteoOpiniones, labels=conteoOpiniones.index, autopct='%1.1f%%', startangle=90)
plt.title('Opiniones con Calificación Máxima por Sitio')
plt.axis('equal')  # Para que el pastel no quede chueco
plt.show()
```
